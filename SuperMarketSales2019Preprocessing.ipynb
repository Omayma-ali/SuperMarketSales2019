# Check List:
- Round all float columns
- Convert 'Time' column datatype to DateTime
- Round minutes in 'Time' column to the nearest hour
- pins 'Time' column
- extract Days from 'Date' column then pins it
- add month name column
- delete 'Invoice ID', 'cogs', 'Branch', 'Unit price','Time', 'day' columns

```python
# rounding all float columns
df['gross income'] = df['gross income'].round(2)
df['gross margin percentage'] = df['gross margin percentage'].round(2)
df['Total'] = df['Total'].round(2)
df['Tax 5%'] = df['Tax 5%'].round(2)
df.sample()
```

```python
# Convert the 'Time' column to datetime
df['Time'] = pd.to_datetime(df['Time'])

# Round the 'Time' column to the nearest hour and extract the hour component
df['Time'] = df['Time'].dt.round('H').dt.hour
```

```python
# Grouping Hours 
df['Hours'] = pd.cut(df['Time'], bins = [0, 3, 6, 9, 12, 15, 18, 21, 24],
                     labels = ['0-3 AM', '3-6 AM', '6-9 AM', '9-12 AM', '12-15 PM', '15-18 PM', '18-21 PM', '21-24 PM'])
df.sample()
```

```python
df['City'].unique()
```

```python
#extract Days & months from 'Date' column
df['day'] = df.index.day
df['month'] = df.index.month_name()
df.sample()
```

```python
df['day'].isna().sum()
```

```python
df['day'] = df['day'].dropna()
```

```python
#Grouping Days
df['Days'] = pd.cut(df['day'], bins=[0,7,14,21,28,32],
                    labels=['week1', 'week2', 'week3', 'week4', '29-31'])
df.sample()
```

```python
df['Days'].isna().sum()
```

```python
df.drop(columns=['Invoice ID', 'cogs', 'Branch', 'Unit price','Time', 'day'],inplace=True)
df.sample()
```
